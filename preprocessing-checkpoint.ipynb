```python
q1=dataframe['age'].quantile(0.25)
q3=dataframe['age'].quantile(0.75)
```

```python
IQR = q3-q1
```

```python
IQR
```

```python
lower_range=q1-1.5*IQR
```

```python
lower_range
```

```python
upper_range=q3+1.5*IQR
```

```python
upper_range
```

```python
trimming_data=dataframe[(dataframe['age']>lower_range) & (dataframe['age']<upper_range) ]
```

```python
trimming_data
```

```python
sns.kdeplot(trimming_data['age'],color='r')
```

```python
sns.histplot(trimming_data['age'],binwidth=5,color='r')
```

```python
sns.boxplot(x=trimming_data['age'])
```

```python
capping_range=dataframe.copy()
```

```python
capping_range['age']=np.where(capping_range['age'] < lower_range,lower_range,
          np.where(capping_range['age'] > upper_range,upper_range,
                   capping_range['age']))
```
